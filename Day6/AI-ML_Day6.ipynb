# Company Name

```
i = input("Enter the email address :\n")
o = ''
a = i.find("@")
b = i.find(".")

for x in range (a+1, b):
  o += i[x]
print (o)
```

# Word sort

```
s = input("Enter the string of words seperated by comma :\n")
l = []
c = ''
out = ''

for x in s:
  if x != ',':
    c += x
  else:
    l.append(c)
    c = ''

l.append(c)
l.sort()

for i in l:
  out = out + i + ','
out = out[:-1]

print(out)
```
