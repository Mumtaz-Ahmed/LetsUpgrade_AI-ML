# Swap two numbers in Python with the help of a temporary variable

```
a = 15
b = 35

print ("Numbers before swap are a =", a, "and b =", b)

temp = a
a = b
b = temp

print ("Numbers after swap are a =", a, "and b =", b)
```

# Swap two numbers in Python without using a temporary variable

```
a = 30
b = 70

print ("Numbers before swap are a =", a, "and b =", b)

a = a+b
b = a-b
a = a-b

print ("Numbers after swap are a =", a, "and b =", b)
```

# Convert fahrenheit to kelvin and celsius both

```
f = float(input("Enter the temp in farenheit:"))

c = (f-32)*5/9
k = c+273.15

print ("Temperature in degree farenheit is ", f)
print ("Temperature in degree celcius is ", c)
print ("Temperature in degree kelvin is ", k)
```

# All the available data types in Python

```
a = 5
b = 6.5
c = 6+5j
d = True
e = "Hello World"
f = (5, 6)
g = [8, 9]
h = {'k1':'v1', 'k2':'v2'}

print ("Datatype of")
print (a, " is", type(a))
print (b, " is", type(b))
print (c, " is", type(c))
print (d, " is", type(d))
print (e, " is", type(e))
print (f, " is", type(f))
print (g, " is", type(g))
print (h, " is", type(h))
```
